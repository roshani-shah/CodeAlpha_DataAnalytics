```python
import pandas as pd
import numpy as np

# Load dataset (using Titanic as example — publicly available)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("=" * 50)
print("📋 EXPLORATORY DATA ANALYSIS")
print("=" * 50)

# 1. Meaningful Questions
print("\n❓ Key Questions:")
print("  - What is the survival rate by gender?")
print("  - Which passenger class had the most survivors?")
print("  - How does age affect survival?")

# 2. Data Structure
print("\n📐 Data Structure:")
print(f"  Shape: {df.shape}")
print(f"\n  Data Types:\n{df.dtypes}")
print(f"\n  First 5 rows:\n{df.head()}")

# 3. Trends, Patterns, Anomalies
print("\n📊 Statistical Summary:")
print(df.describe())

print("\n🔍 Missing Values:")
missing = df.isnull().sum()
print(missing[missing > 0])

print("\n📈 Survival Rate by Gender:")
print(df.groupby("Sex")["Survived"].mean().round(2))

print("\n📈 Survival Rate by Class:")
print(df.groupby("Pclass")["Survived"].mean().round(2))

# 4. Hypothesis Testing
from scipy import stats
male_survival = df[df["Sex"] == "male"]["Survived"]
female_survival = df[df["Sex"] == "female"]["Survived"]
t_stat, p_value = stats.ttest_ind(male_survival, female_survival)
print(f"\n🧪 Hypothesis Test (Gender vs Survival):")
print(f"  T-statistic: {t_stat:.4f}, P-value: {p_value:.6f}")
print(f"  {'✅ Significant difference' if p_value < 0.05 else '❌ No significant difference'}")

# 5. Data Issues
print("\n⚠️ Data Issues Detected:")
print(f"  - 'Cabin' has {df['Cabin'].isnull().sum()} missing values ({df['Cabin'].isnull().mean()*100:.1f}%)")
print(f"  - 'Age' has {df['Age'].isnull().sum()} missing values")
print(f"  - Duplicates: {df.duplicated().sum()}")
```
